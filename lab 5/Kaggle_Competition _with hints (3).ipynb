```python
# drop the column 'glass' for the training dataset since it has too many NULLs
# *** TODO ***
df_train.drop("glass", axis=1,inplace=True)
# drop the column 'glass' for the test dataset (to match the training)
df_test.drop("glass", axis=1,inplace=True)
```

```python
# isOrganic is a binary value, lets convert it to 0/1 for the training dataset
# *** TODO ***
df_train.isOrganic = df_train["isOrganic"].str.replace("Y","1")
df_train.isOrganic = df_train["isOrganic"].str.replace("N","0")

# isOrganic is a binary value, lets convert it to 0/1 for the test dataset
# *** TODO ***
df_test.isOrganic = df_test["isOrganic"].str.replace("Y","1")
df_test.isOrganic = df_test["isOrganic"].str.replace("N","0")
```

```python
# Lets examine the column 'available'. Note that it is a categorical variable
df_train["available"].value_counts()
```

```python
# Convert column 'available' using oneHotEncoding/get_dummies

df_train = pd.concat([df_train,pd.get_dummies(df_train['available'], prefix='available',dummy_na=True)],axis=1).drop(['available'],axis=1)
df_train.head(2)
```

```python
# It probably makes sense to rename those columns to something shorter. 
# Look up how to rename columns (inplace) in a dataframe
# *** TODO ***
df_train = df_train.rename(columns={"available_Available at the same time of year, every year.":"avaiable same time every year"})
df_train = df_train.rename(columns={"available_Available during the fall months.":"avaiable fall"})
df_train = df_train.rename(columns={"available_Available during the spring months.":"avaiable spring"})
df_train = df_train.rename(columns={"available_Available during the summer months.":"avaiable summer"})
df_train = df_train.rename(columns={"available_Available during the winter months.":"avaiable winter"})
df_train = df_train.rename(columns={"available_Available year round as a staple beer.":"avaiable year around"})
df_train = df_train.rename(columns={"available_Beer is not available.":"not avaiable"})
df_train = df_train.rename(columns={"available_Limited availability.":"limited availability"})

# Do the same transformations for the df_test
# *** TODO ***
                         
df_test = df_test.rename(columns={"available_Available at the same time of year, every year.":"avaiable same time every year"})
df_test = df_test.rename(columns={"available_Available during the fall months.":"avaiable fall"})
df_test = df_test.rename(columns={"available_Available during the spring months.":"avaiable spring"})
df_test = df_test.rename(columns={"available_Available during the summer months.":"avaiable summer"})
df_test = df_test.rename(columns={"available_Available during the winter months.":"avaiable winter"})
df_test = df_test.rename(columns={"available_Available year round as a staple beer.":"avaiable year around"})
df_test = df_test.rename(columns={"available_Beer is not available.":"not avaiable"})
df_test = df_test.rename(columns={"available_Limited availability.":"limited availability"})
df_train
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import numpy as np
pipeline = make_pipeline(
    StandardScaler(),
    LinearRegression()
)

pipeline.fit(X=X_train, y=y_train)

scores = cross_val_score(pipeline, X=X_train, y=y_train, cv=10)

print(np.mean(scores))
```
