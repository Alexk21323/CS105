```python
import pandas as pd 
from pandas.io.json import json_normalize

import requests
reponse = requests.get("https://api.weather.gov/points/33.9763,-117.3567/forecast")
data = reponse.json()
data['properties']['periods']
```

Gather weather information for the city Riverside for the next 2 weeks 

```python
import numpy as np

df = json_normalize(data['properties']['periods'])
df['temperature'] = df['temperature'].astype(str)+ ' F'
df.index = np.arange(1, len(df)+1)
df.drop(columns=['number', 'icon', 'temperatureUnit'])
```
